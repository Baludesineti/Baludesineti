# **Task 1.3: Data Cleaning**

## Detects and handles missing values in the dataset to ensure data integrity.

```python
# Fill missing values in numerical columns with the mean.
df['Train_No'].fillna(df['Train_No'].mean(), inplace=True)
```

```python
# Fill missing values in categorical columns with the most frequent value.
df['Source_Station_Name'].fillna(df['Source_Station_Name'].mode()[0], inplace=True)
df['Destination_Station_Name'].fillna(df['Destination_Station_Name'].mode()[0], inplace=True)
```

```python
# Verify if there are still any missing values
remaining_missing = df.isnull().sum()
print(remaining_missing)
```

# Standardise the format of station names for consistency, for instance, converting all names to uppercase.

```python
# Convert station names to uppercase
df['Source_Station_Name'] = df['Source_Station_Name'].str.upper()
df['Destination_Station_Name'] = df['Destination_Station_Name'].str.upper()
df
```

# **Task 2.3: Data Enrichment**

# Enrich the dataset by adding a new column that categorises trains based on their operating days (e.g., 'Weekday', 'Weekend').

```python
# Define a function to categorize trains based on operating days
def categorize_train(days):
  if "Saturday" in days or "Sunday" in days:
    return "Weekend"
  else:
    return "Weekday"

# Apply the function to create a new column 'Train_Category'
df['Train_Category'] = df['days'].apply(categorize_train)

# Display the updated DataFrame
print(df)
```

# **Level 3: Advanced Data Analysis**

# **Task 3.1: Pattern Analysis**

# Analyse the distribution of train journeys throughout the week and visualise the findings using bar plots or histograms.

```python
# Group the data by day of the week and count the number of trains
trains_per_day = df['days'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
trains_per_day.plot(kind='bar')
plt.title('Distribution of Train Journeys Throughout the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trains')
plt.xticks(rotation=45)
plt.show()
```

# Identify and examine patterns or trends in train operations based on source and destination stations.

```python
# Group by source and destination stations and count the frequency
route_freq = df.groupby(['Source_Station_Name', 'Destination_Station_Name']).size().reset_index(name='frequency')

# Sort by frequency
route_freq = route_freq.sort_values(by='frequency', ascending=False)

# Display the most common routes
print(route_freq.head(10))
```

```python
# Visualize the top 10 most common routes
top_routes = route_freq.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='frequency', y='Source_Station_Name', hue='Destination_Station_Name', data=top_routes, dodge=False)
plt.title('Top 10 Most Common Train Routes')
plt.xlabel('Frequency')
plt.ylabel('Routes')
plt.show()
```

# **Task 3.2: Correlation and Insights**

# Investigate any correlation between the number of trains and specific days of the week.

```python
# Create a contingency table of days of the week and train counts
contingency_table = pd.crosstab(df['days'], df['Train_No'])

# Calculate the Chi-squared statistic and p-value
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-squared statistic:", chi2)
print("P-value:", p)

# Interpret the results
if p < 0.05:
  print("There is a statistically significant correlation between the number of trains and specific days of the week.")
else:
  print("There is no statistically significant correlation between the number of trains and specific days of the week.")
```

# Create various visualisations (bar charts, line charts, heatmaps) to illustrate different aspects of the data, such as the number of trains per station and day-wise distribution.

```python
# Bar chart for number of trains per station
plt.figure(figsize=(12, 6))
df['Source_Station_Name'].value_counts().plot(kind='bar')
plt.title('Number of Trains per Station')
plt.xlabel('Station')
plt.ylabel('Number of Trains')
plt.xticks(rotation=90)
plt.show()

# Line chart for day-wise distribution of trains
plt.figure(figsize=(10, 6))
df['days'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Day-wise Distribution of Trains')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trains')
plt.grid(True)
plt.show()
```

```python
# Bar chart for number of trains per station using Seaborn
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Source_Station_Name', order=df['Source_Station_Name'].value_counts().index)
plt.title('Number of Trains per Station')
plt.xlabel('Station')
plt.ylabel('Number of Trains')
plt.xticks(rotation=90)
plt.show()

# Line chart for day-wise distribution of trains using Plotly
import plotly.express as px
day_counts = df['days'].value_counts().sort_index()
fig = px.line(x=day_counts.index, y=day_counts.values, markers=True)
fig.update_layout(title='Day-wise Distribution of Trains', xaxis_title='Day of the Week', yaxis_title='Number of Trains')
fig.show()
```

# **Task 4.2: Reporting**

# Compile a comprehensive report summarising all findings, visualisations, and insights derived from the data.

```python
# Analyze the most common source and destination stations
source_station_counts = df['Source_Station_Name'].value_counts()
destination_station_counts = df['Destination_Station_Name'].value_counts()

# Display the top 5 most common source and destination stations
top_source_stations = source_station_counts.head(5)
top_destination_stations = destination_station_counts.head(5)

# Plot the top 5 source stations
plt.figure(figsize=(10, 6))
sns.barplot(x=top_source_stations.values, y=top_source_stations.index, palette="viridis")
plt.title('Top 5 Source Stations')
plt.xlabel('Number of Trains')
plt.ylabel('Source Station')
plt.savefig('top_source_stations.png')
plt.show()
```

```python
# Plot the top 5 destination stations
plt.figure(figsize=(10, 6))
sns.barplot(x=top_destination_stations.values, y=top_destination_stations.index, palette="viridis")
plt.title('Top 5 Destination Stations')
plt.xlabel('Number of Trains')
plt.ylabel('Destination Station')
plt.savefig('top_destination_stations.png')
plt.show()
```

```python
# Clean the 'days' column to remove anomalies
df['days'] = df['days'].str.replace('d', '')

# Analyze the frequency of trains operating on different days
days_counts = df['days'].value_counts()

# Plot the frequency of trains operating on different days
plt.figure(figsize=(10, 6))
sns.barplot(x=days_counts.index, y=days_counts.values, palette="viridis")
plt.title('Frequency of Trains Operating on Different Days')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trains')
plt.savefig('train_frequency_by_day.png')
plt.show()
```

```python
# Analyze connectivity between key stations
connectivity_counts = df.groupby(['Source_Station_Name', 'Destination_Station_Name']).size().reset_index(name='train_count')

# Sort the DataFrame by train_count in descending order to get the most common routes
top_connectivity = connectivity_counts.sort_values(by='train_count', ascending=False).head(10)

# Save the top connectivity to a CSV file for the report
top_connectivity.to_csv('top_connectivity.csv', index=False)

# Generate a comprehensive report
report = f"""
# Railway Data Analysis Report

# 1. Top 5 Source Stations
The following are the top 5 source stations based on the number of trains:
{top_source_stations.to_string()}

![Top Source Stations](top_source_stations.png)

# 2. Top 5 Destination Stations
The following are the top 5 destination stations based on the number of trains:
{top_destination_stations.to_string()}

![Top Destination Stations](top_destination_stations.png)

# 3. Frequency of Trains Operating on Different Days
The frequency of trains operating on different days of the week is as follows:
{days_counts.to_string()}



# 4. Top 10 Train Routes

{top_connectivity.to_string(index=False)}

The detailed data can be found in the 'top_connectivity.csv' file.
"""
```
